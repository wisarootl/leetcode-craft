# 790. Domino and Tromino Tiling

https://leetcode.com/problems/domino-and-tromino-tiling/

```python
class Solution:
    # time = O(n)
    # space = O(n)
    def numTilings(self, n: int) -> int:
        dp = [1, 2, 5] + [0] * n
        for i in range(3, n):
            dp[i] = (dp[i - 1] * 2 + dp[i - 3]) % 1000000007
        return dp[n - 1]
```
